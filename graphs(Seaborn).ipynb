```python
#1. scatterplot
sns.scatterplot(d ata=tips, x="total_bill", y="tip", hue="time", size="size", palette="deep")
plt.title("Scatterplot of total bills vs tips")
```

```python
#line plot

sns.lineplot(data=tips, x='size', y='total_bill', hue='sex',ci=None, markers='o')
plt.title("lineplot of total bill vs sixe")
```

```python
# bar plot
sns.barplot(data=tips, x='day',y='total_bill', hue='sex', palette='muted')
plt.title("barplot of total bill by day")
```

```python
tips.columns
```

```python
#boxplot
sns.boxplot(data=tips, x='day', y='tip', hue='smoker', palette='pastel')
plt.title("Boxplot of tips by day and smoker status")
```

```python
# violin plot

sns.violinplot(data=tips, x='day', y='total_bill', hue='time', split=True, palette='dark')
plt.title("violin plot of total bill by day and time")
```

```python
tips.columns
```

```python
# count plot
sns.countplot(data=tips, x='day', hue='smoker', palette='colorblind')
plt.title("count plot of days by smoker status")
```

```python
tips.columns
```

```python
#regression plot
sns.regplot(data=tips, x='total_bill', y='tip', scatter_kws={'s':50}, line_kws={'color':'black'})
plt.title("regression plot of total bill vs tip")
plt.show()
```

```python
# histogram
sns.histplot(data=tips, x='total_bill', bins=20, kde=True, color='pink')
plt.title("histogram of total bill with kde")
```

```python
# pairplot
sns.pairplot(data=tips, hue='sex', vars=['total_bill','tip','size'],palette='husl')
plt.suptitle("pair plot: numberic variable by gender", y=1.02)
 
```

```python
# catplot
sns.catplot(data=tips, x='day', y='tip' , hue='sex', kind='point', palette='bright')
plt.title("catplot(points): tips by day and gender")
```

```python
# joint plot
sns.jointplot(data=tips, x='total_bill',y='tip',kind='scatter', hue='smoker', color='pink', palette='coolwarm')
plt.suptitle("joinplot: total bills vs tips", y=1.02)
```

```python
# Facetgrid
g=sns.FacetGrid(tips, col='time', row='smoker', margin_titles=True).map(sns.histplot,'total_bill')
```

```python
# strip plot
sns.stripplot(data=tips, x='day' , y='tip' , hue='sex', jitter=True, palette='Set1')
plt.title("strip plot: tips by day and gender")
```

```python
# KDE plot
sns.kdeplot(data=tips, x='total_bill',hue='sex', fill=True, palette="tab10")
plt.title("kde plot: total bill density by gender")
```
